```python
import sklearn
from sklearn.preprocessing import OneHotEncoder
```

```python
import numpy as np
```

```python
onehotencoder = OneHotEncoder()
```

```python
import pandas as pd
```

```python
data = pd.read_csv('ImageCLEF2013PlantTask/info/mapped_classes_by_folder.csv')
data
```

```python
X = onehotencoder.fit_transform(data.FolderName.values.reshape(-1,1)).toarray()
```

```python
X[1]
```

```python
dfOneHot = pd.DataFrame(X, columns = ["species_"+str(int(i)) for i in range(data.shape[0])]) 
dfOneHot
```

```python
df = pd.concat([data, dfOneHot], axis=1)
df= df.drop(['Species'], axis=1)
df= df.drop(['FolderName'], axis=1)
df
```

```python
df.to_csv('one_hot_encoded_labels.csv')
```
